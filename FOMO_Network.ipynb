```python
import os
import shutil
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])
```

```python
dir = "/content/tiny_fruit/JPEGImages"
output_dir = '/content/fruits_dataset'

# Create output directories if they don't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
# Iterate through the images in the directory
for filename in os.listdir(dir):
    if 'apple' in filename.lower():
        class_dir = os.path.join(output_dir, 'apple')
    elif 'orange' in filename.lower():
        class_dir = os.path.join(output_dir, 'orange')
    elif 'banana' in filename.lower():
        class_dir = os.path.join(output_dir, 'banana')
    else:
        class_dir = os.path.join(output_dir, 'others')

    # Create the class directory if it doesn't exist
    if not os.path.exists(class_dir):
        os.makedirs(class_dir)

    # Move the image to the corresponding class directory
    src_path = os.path.join(dir, filename)
    dst_path = os.path.join(class_dir, filename)
    shutil.move(src_path, dst_path)
    print(f'Moving {filename} to {class_dir}')

print("Images have been organized into class-based directories.")
```

```python
#Neural Network architecture
class NeuralNetwork(nn.Module):
    def __init__(self, base_model, num_classes):
        super(NeuralNetwork, self).__init__()
        self.base_model = base_model

        # Freeze base model parameters
        for parameter in self.base_model.parameters():
            parameter.requires_grad = False

        self.base_model.classifier = nn.Sequential(
            nn.Linear(self.base_model.last_channel, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )


    def forward(self, x):
        x = self.base_model(x)
        return x
```

```python
#Neural Network with FOMO architecture
class FOMO_Network(nn.Module):
    def __init__(self, num_classes):
        super(FOMO_Network, self).__init__()

        # Load the MobileNetV2 model pre-trained on ImageNet
        model_head = torchvision.models.mobilenet_v2(pretrained=True)

        # Freeze the model parameters
        for param in model_head.parameters():
            param.requires_grad = False

        # Cut the model at the block_6_expand_relu layer
        self.features = nn.Sequential(*list(model_head.features.children())[:7])

        # Get the last layer
        #last_layer = features[-1]
        last_layer = list(self.features.children())[-1]

        # Get the number of output channels of the last layer
        last_channel = last_layer.out_channels

        # Add the additional head
        self.head = nn.Sequential(
            nn.Conv2d(last_channel, 32, kernel_size=1, stride=1),
            nn.ReLU(),
            nn.Dropout(0.5)
        )

        # Add global average pooling
        self.pool = nn.AdaptiveAvgPool2d((1, 1))

        # Add the final classifier layer
        self.logits = nn.Conv2d(32, num_classes, kernel_size=1, stride=1)

    def forward(self, x):
        x = self.features(x)
        #print(f'After features: {x.shape}')
        x = self.head(x)
        #print(f'After Head: {x.shape}')
        x = self.pool(x)
        #print(f'After Pool: {x.shape}')
        # Final classifier layer
        x = self.logits(x)
        #print(f'After Logits: {x.shape}')
        x = x.view(x.size(0), -1)  # Flatten the output for classification
        #print(f'After Flatten: {x.shape}')
        return x
```

```python
num_classes = len(train_dataset.dataset.classes)
model = FOMO_Network(num_classes)
print(model)
```

```python
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
#Training loop
if torch.cuda.is_available():
    # Move the model to the GPU if available
    #model = model.cuda()
    num_epochs = 25
else:
    num_epochs = 2

#num_epochs = 25

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in iter(train_loader):
        """if torch.cuda.is_available():
          inputs, labels = inputs.cuda(), labels.cuda()"""
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
```

```python
# Test loop
model.eval()
test_loss = 0.0
correct = 0
with torch.no_grad():
    for inputs, labels in test_loader:
      """if torch.cuda.is_available():
        inputs, labels = inputs.cuda(), labels.cuda()"""
      outputs = model(inputs)
      _loss = criterion(outputs, labels)
      test_loss += _loss.item() * inputs.size(0)
      _, preds = torch.max(outputs, 1)
      correct += torch.sum(preds == labels.data)

test_loss /= len(test_loader.dataset)
test_accuracy = correct.double() / len(test_loader.dataset)
print(f'Test Loss: {test_loss:.4f}, Accuracy: {test_accuracy:.4f}')
```

```python
#Checking Model Size

def model_summary(model):
    total_params = sum(p.numel() for p in model.parameters())
    param_size = sum(p.nelement() * p.element_size() for p in model.parameters())
    buffer_size = sum(b.nelement() * b.element_size() for b in model.buffers())

    size_all_mb = (param_size + buffer_size) / 1024**2
    return total_params, size_all_mb

total_params, size_all_mb = model_summary(model)
print(f'Total number of parameters: {total_params}')
print(f'Model size: {size_all_mb:.2f} MB')
```

```python
from PIL import Image

#Function to load and preprocess the image
def load_and_preprocess_image(image_path):
    # Load the image
    image = Image.open(image_path).convert('RGB')

    # Apply transformations
    image = transform(image)

    # Add a batch dimension
    image = image.unsqueeze(0)  # Shape: [1, 3, 224, 224]

    """# Move the image to the GPU if available
    if torch.cuda.is_available():
        image = image.cuda()"""

    return image
```

```python
# Function to get the model's prediction
def predict(image_path, model):
    # Preprocess the image
    image = load_and_preprocess_image(image_path)

    # Set the model to evaluation mode
    model.eval()

    # Disable gradient calculation
    with torch.no_grad():
        # Pass the image through the model
        output = model(image)

    # Get the predicted class
    _, predicted_class = torch.max(output, 1)

    return predicted_class.item(), output
```
